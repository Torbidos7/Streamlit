```python
import pandas as pd
from pycaret.classification import *
```

```python
df = pd.read_csv('./data/winequality-red.csv', sep=';')
```

Dataset visualization and description


```python
df.describe()
```

```python
df.quality.value_counts()
```

```python
session_1 = setup(data=df, target='quality', session_id=123, log_experiment=False, experiment_name='wine1')
```

Model available

```python
models()
```

Model comparison

```python
best_model = compare_models()
results = pull()
```

```python
print(results)
```

```python
results[:3]
```

Tuning the top 3 modes

```python
models = []
models.append( results[:3].index.tolist())
```

```python
models[0][0]
```

```python
var1 = create_model(str(models[0][0]))
```

```python
tuned_var1 = tune_model(var1)
```

Plotting ROC

```python
plot_model(tuned_var1, plot='confusion_matrix')
plot_model(tuned_var1, plot='auc')
```
